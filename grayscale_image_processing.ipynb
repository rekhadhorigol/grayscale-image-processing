#PROJECT 2 – GRAYSCALE IMAGE OPERATIONS

Objective: To apply basic image processing techniques on grayscale images using NumPy

Concepts Used:
- Pixel manipulation
- Image transformation
- Array operations

Outcome: Shows how grayscale images can be processed using matrix operations

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from google.colab import files
```

```python
#Task 1
#Upload image & convert to grayscale
uploaded = files.upload()

filename = list(uploaded.keys())[0]
im = Image.open(filename).convert('L')   # convert to grayscale
ImJPG = np.array(im)
```

```python
import matplotlib.pyplot as plt
from PIL import Image

# Reload the original image without converting to grayscale
im_color = Image.open(filename)
ImJPG_color = np.array(im_color)

plt.imshow(ImJPG_color)
plt.axis('off')
plt.title("Original Colored Image")
plt.show()
```

```python
#Task 2
#Find dimensions of the image
m, n = ImJPG.shape
print("Rows:", m)
print("Columns:", n)
```

```python
#Task 3
#Check whether image pixels are integers
print(np.issubdtype(ImJPG.dtype, np.integer))
```

```python
#Task 4
#Find maximum and minimum pixel intensity
print("Maximum pixel value:", np.max(ImJPG))
print("Minimum pixel value:", np.min(ImJPG))
```

```python
#Task 5
#Display the grayscale image
plt.imshow(ImJPG, cmap='gray')
plt.axis('off')
plt.title("Grayscale Image")
plt.show()
```

```python
#Task 6
#Crop the image
crop = ImJPG[100:m-100, 100:n-70]

plt.imshow(crop, cmap='gray')
plt.axis('off')
plt.title("Cropped Image")
plt.show()
```

```python
#Task 7
#Paste cropped image into a black background
border = np.zeros((m, n), dtype=np.uint8)
border[100:m-100, 100:n-70] = crop

plt.imshow(border, cmap='gray')
plt.axis('off')
plt.title("Image with Border")
plt.show()
```

```python
#Task 8
#Vertical flip of image
plt.imshow(np.flipud(ImJPG), cmap='gray')
plt.axis('off')
plt.title("Vertical Flip")
plt.show()
```

```python
#Task 9
#Effect of transpose on an image
plt.imshow(ImJPG.T, cmap='gray')
plt.axis('off')
plt.title("Transposed Image")
plt.show()
```

Transposing an image swaps rows and columns, resulting in rotation and mirroring of the image.

```python
#Task 10
#Horizontal flip of image
plt.imshow(np.fliplr(ImJPG), cmap='gray')
plt.axis('off')
plt.title("Horizontal Flip")
plt.show()
```

```python
#Task 11
#Rotation using rot90
plt.imshow(np.rot90(ImJPG), cmap='gray')
plt.axis('off')
plt.title("90 Degree Rotation")
plt.show()
```

rot90() rotates the image 90° anticlockwise.

```python
#Task 12
#Invert grayscale image
invert = 255 - ImJPG

plt.imshow(invert, cmap='gray')
plt.axis('off')
plt.title("Inverted Image")
plt.show()
```

```python
#Task 13
#Brighten / Darken the image
bright = np.clip(ImJPG + 30, 0, 255)

plt.imshow(bright, cmap='gray')
plt.axis('off')
plt.title("Brightened Image")
plt.show()
```

```python
#Task 14
#Andy Warhol–style effect
warhol = np.hstack((ImJPG, 255 - ImJPG))

plt.imshow(warhol, cmap='gray')
plt.axis('off')
plt.title("Andy Warhol Effect")
plt.show()
```

```python
#Task 15
#Convert grayscale image to black & white
bw = np.uint8(255 * np.floor(ImJPG / 128))

plt.imshow(bw, cmap='gray')
plt.axis('off')
plt.title("Black and White Image")
plt.show()
```

```python
#Task 16
#Reduce grayscale levels (8 levels)
ImJPG8 = np.uint8(np.round(ImJPG / 255 * 7) * (255 / 7))

plt.imshow(ImJPG8, cmap='gray')
plt.axis('off')
plt.title("8-Level Grayscale Image")
plt.show()
```

```python
#Task 17
#Change contrast using scalar multiplication
contrast = np.clip(ImJPG * 1.25, 0, 255).astype(np.uint8)

plt.imshow(contrast, cmap='gray')
plt.axis('off')
plt.title("Contrast Enhanced Image")
plt.show()
```

```python
#Task 18
#Gamma correction
gamma = 1.1
gamma_corrected = np.uint8((ImJPG / 255) ** gamma * 255)

plt.imshow(gamma_corrected, cmap='gray')
plt.axis('off')
plt.title("Gamma Corrected Image")
plt.show()
```
